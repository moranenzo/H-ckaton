# Load data

```python
from pandas import DataFrame, read_csv

# Function to load data
def load_data(
        data_file: str,
        n_rows: int | None = None,
) -> DataFrame:
    return read_csv(data_file, index_col='row_index', nrows=n_rows)
```

```python
import pandas as pd

#train_data = file_manager.load_training_data()

# load datsets
train_file = '../X_train.csv'
test_file = '../X_test.csv'
n_rows = 100_000

test_data = load_data(test_file)
# load reduced training dataset
tiny_train_data = load_data(train_file, n_rows=n_rows)
```

# Split raw data by variable category (piezo, weather, ...)

We are only interested in training data.

```python
from pandas import DataFrame

# Function to extract variable category
def extract_category(
        df: DataFrame,
        category: str,
) -> DataFrame:
    return df[df.columns[df.columns.str.startswith(category)]]
```

```python
# Example with meteo variables
withdrawal_variables = extract_category(tiny_train_data, 'prelev')
withdrawal_variables.head()
```

# Withdrawal Information

```python
# Display withdrawal features information
withdrawal_variables.info()
    
```

```python
# Create a copy to avoid modifying a slice of the DataFrame
withdrawal_variables = withdrawal_variables.copy()

# Loop to drop the specified columns
for i in range(3):
    columns_to_drop = [
        f"prelev_structure_code_{i}",
        f"prelev_usage_label_{i}",
        f"prelev_volume_obtention_mode_label_{i}",
        f"prelev_longitude_{i}",
        f"prelev_latitude_{i}",
        f"prelev_commune_code_insee_{i}"
    ]
    withdrawal_variables.drop(columns=columns_to_drop, inplace=True, axis=1)

# Handle NaN values for specific columns
cols_to_merge = ["prelev_volume_0", "prelev_volume_1", "prelev_volume_2", "prelev_other_volume_sum"]
withdrawal_variables[cols_to_merge] = withdrawal_variables[cols_to_merge].fillna(0)

# Create a new column for the total volume
withdrawal_variables["prelev_volume_total"] = withdrawal_variables[cols_to_merge].sum(axis=1)

# Drop the individual volume columns
withdrawal_variables.drop(columns=cols_to_merge, inplace=True)
```
